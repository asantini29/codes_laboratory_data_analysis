# Build a model explaining the emission of galaxies in the optical band 

## Assumptions:  

1) the emission is only due to stars

2) stars are perfect black bodies

3) IMF: $\xi(M) \propto M^{-2.35}$

## Functions

```python
#return the maximum mass as a function of the age of the stellar population
def max_mass(t, M_sun):
    t_sun = 1e10 * u.yr #10 Gy
    return M_sun * ((t_sun / t).value)**0.5
```

```python
#scale factor and temperature
def scale_and_temperature(M, M_sun):
    L_sun = const.L_sun.cgs.value
    R_sun = const.R_sun.cgs.value
    sigma = const.sigma_sb.cgs.value
    eta = 3.5 #exponent of the luminosity
    
    if (M >= 1.3 * M_sun): 
        xi = 0.56 #cno cycle
    else:
        xi = 0.79 #pp chain
        
    R = R_sun * (M/M_sun)**xi
    L = L_sun * (M/M_sun)**eta
    
    scale_factor = (M/M_sun)**(-2.35) * R**2 
    T = (L / (4 * np.pi * R**2 * sigma))**(1/4)
    
    return scale_factor, T
```

## Main

```python
M_sun = const.M_sun.cgs.value
M_min = 0.1 * M_sun

#wavelenght array
wave = np.linspace(3000, 8000, 5000) * u.AA
```

```python
#masses
age = 1e8 * u.yr  #age of the distribution
M_max = max_mass(age, M_sun)
M_array = np.linspace(M_min, M_max, int(10*M_max / M_sun))
```

```python
scale_factor_array, temp_array = np.hsplit(np.array([scale_and_temperature(M, M_sun) for M in M_array]), 2)
```

```python
plt.plot(temp_array, M_array / M_sun)
plt.axhline(1, c= 'k', ls = '--', label = 'Sun')
plt.vlines(5778, M_array.min() / M_sun, M_array.max() / M_sun, colors = 'k', ls = '--')
plt.ylabel("$M [M_\odot]$")
plt.xlabel("$T [K]$")
plt.title("Mass vs temperature (T increasing from right to left)")
#plt.xscale("log")
plt.xlim(max(temp_array), min(temp_array))
plt.loglog()
plt.legend(frameon = False)
plt.show()
```

```python
bb = np.sum([models.BlackBody(temperature = T * u.K, scale = scale_factor * u.erg / (u.cm * u.cm * u.AA * u.s * u.sr)) for T,scale_factor in zip(temp_array, scale_factor_array)])
```

```python
flux = bb(wave)
with quantity_support():
    plt.figure()
    plt.plot(wave, flux / bb(5000 * u.AA)) #normalize to F(5000 A)
    #plt.plot(wave, flux) #normalize to F(5000 A)    
    plt.ylabel("$F(\lambda) / F(5000 \, \dot{A})$")
    plt.title("Summed spectrum")
    plt.show()
```
